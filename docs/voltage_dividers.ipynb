```python
lt = ltparser.LTspice()
lt.read("../tests/examples/voltage_divider_1.asc")
lt.make_netlist()
print(lt.netlist)
cct = lt.circuit()
cct.draw(scale=0.25)
v_R2 = cct.R2.V
print("The DC voltage across R2 is %.2fV" % v_R2.dc.fval)
```

```python
lt = ltparser.LTspice()
lt.read("../tests/examples/voltage_divider_1.asc")
lt.make_netlist(minimal=True)
print(lt.netlist)
cct = lt.circuit()
vA = cct.P1.V
vB = cct.P2.V
print("The DC voltage at A is %.2fV" % vA.dc.fval)
print("The DC voltage at B is %.2fV" % vB.dc.fval)
```
